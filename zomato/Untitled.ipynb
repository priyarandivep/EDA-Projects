```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Dataset

```python
df= pd.read_excel("zomato.xlsx")
df
```

```python
country= pd.read_excel("Country-Code.xlsx")
```

# Merging df and country Dataframe

```python
#Joining Two file
```

```python
zomato= pd.merge(df, country,how="inner", on="Country Code")
```

# 1. Top 5 countries which has highest number of Restaurants

```python
a= zomato.groupby("Country")["Restaurant Name"].count().sort_values(ascending= False)
a
```

```python
plt.figure(figsize= (8,5))
sns.barplot(x= a.index[:5] , y=a.values[:5])
```

# Observations:
    India has highest number of restarants. Roughly 90 % of restaurants are only in India.
    

# 2. Number of restaurants falling in different Rating range

```python
b= zomato.groupby(["Rating text"])["Restaurant Name"].count()
b
```

```python
plt.figure(figsize= (10,8))
sns.barplot(x= b.index, y= b.values )
```

##### Observations:
    Most of the customers has given average ratings to the Restaurant.
    Some customers has not given any ratings to the Restaurant.
    

# 3.  India Restaurant Count with respect to their  the corresponding rating band

```python
c= zomato[zomato["Country"]=="India"].groupby("Rating text")["Restaurant Name"].count()
c
```

```python
sns.barplot(x= c.index, y=c.values)
plt.xticks(rotation= "vertical")
plt.show()
```

##### Obervations:
    In India Most of the Resaturant  has Average rating.
