### 20. Classify Movies Based on Ratings [Excellent, Good and Average]

```python
def rating_cat(x):
    if x>= 7:
        return "Excellent"
    elif x<7.0 and x>=5.0:
        return "Good"
    else:
        return "Average"
        
```

```python
df["Rating_cat"]= df["Rating"].map(rating_cat)
```

### 22. Find Unique Values From Genre

```python
a= df["Genre"]
a
```

```python
 c=[]
for i in a.values:
   
    d=i.split(",")
    c.append(d)
```

```python
c
```

```python
un= []
for i in c:
    for it in i:
        if it not in un:
            un.append(it)
        
```
